```python
import pandas as pd
import numpy as np
import re
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_csv('houses.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.sample(5)
```

```python
df.drop(columns = ['link', 'property_id'], inplace = True)
```

```python
df.sample(5)
```

```python
df['society'].value_counts()
```

```python
df['society'].value_counts().shape
```

```python
df.loc[:, 'society'] = df['society'].apply(lambda x: re.sub(re.compile(r'\d+\.\d+ ★'), '', str(x))).str.strip().str.lower()
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df['society'].value_counts().shape
```

```python
df.sample(2)
```

```python
df['price'].str.split().str.get(1).value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price (x) :
  if type(x) == float :
    return x
  if x[1] == 'Lac' :
    return round(float(x[0])/100,2)
  else :
    return round(float(x[0]), 2)
```

```python
df.loc[:, 'price'] = df['price'].str.split().apply(treat_price)
```

```python
df.sample(3)
```

```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
df['price_per_sqft'].value_counts()
```

```python
df['price_per_sqft'].str.split('/').str.get(1).value_counts()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹', '').str.replace(',', '').str.strip().astype('float')
```

```python
df.sample(3)
```

```python
df['bedRoom'].isnull().sum()
```

```python
df[df['bedRoom'].isnull()]
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'] = df['bedRoom'].apply(lambda x : re.sub(re.compile(r'Bedrooms?'), '', str(x))).str.strip().astype('int')
```

```python
df['bedRoom'].value_counts()
```

```python
df.sample(3)
```

```python
df['bathroom'].isnull().sum()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'] = df['bathroom'].apply(lambda x : re.sub(r'Bathrooms?', '', x)).str.strip().astype('int')
```

```python
df.sample(3)
```

```python
df['balcony'].isnull().sum()
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'] = df['balcony'].str.split().str.get(0).str.replace('No', '0')
```

```python
df.sample(3)
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].fillna('not available', inplace = True)
df['additionalRoom'] = df['additionalRoom'].str.lower()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['noOfFloor'] = df['noOfFloor'].str.get(0)
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df.sample(3)
```

```python
df.fillna({'facing' : 'NA'}, inplace = True)
```

```python
df['area'] = round((df['price']*10000000)/df['price_per_sqft'])
```

```python
df.insert(loc=1,column='property_type',value='house')
```

```python
df.shape
```

```python
df['price'] = df['price'].astype('float')
```

```python
df.info()
```

```python
df.to_csv('houses_cleaned.csv')
```
