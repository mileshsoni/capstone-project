```python
def treat_price (x) :
  if type(x) == float :
    return x
  if x[1] == 'Lac' :
    return round(float(x[0])/100,2)
  else :
    return round(float(x[0]), 2)
```

```python
df['price'] = df['price'].str.split().apply(treat_price)
```

```python
def handle_floor (x) :
  if type(x) == float :
    return x
  if (x == 'Lower') | (x == 'Ground'):
    return '0'
  if x == 'Basement' :
    return '-1'
  # extract only digit
  pattern = re.compile(r'\d+')
  return re.findall(pattern, x)[0]
```

```python
df['floorNum'] = df['floorNum'].str.split().str.get(0).apply(handle_floor)
```
